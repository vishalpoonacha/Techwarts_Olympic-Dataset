```python
population=pd.read_excel('/content/Population data Assignment 1 Predict This.xlsx')
population
```

```python
population.to_csv ("population.csv", 
                  index = None,
                  header=True)
Population = pd.DataFrame(pd.read_csv("population.csv"))
Population
```

```python
Population.rename(columns={'Country': 'Team/NOC'}, inplace=True)
Population
```

```python
Population['Team/NOC'] = Population['Team/NOC'].str.upper()
Population
```

```python
medals_population=pd.merge(medals,Population,on="Team/NOC")
medals_population
```
